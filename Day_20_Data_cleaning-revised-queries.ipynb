```python
def missingValuesPercentage(df):
  print(f"{'Column':<25} {'missingCount':<10}  {'missingPercentage'} ")
  print("-----------------------------------------------------------")
  for col in df.columns:
    missingCount = df[col].isnull().sum()
    missingPercentage = (missingCount / len(df)) * 100
    print(f"{col:<25}  {str(missingCount):<10}     {str(missingPercentage)}{'%'} ")
```

```python
missingValuesPercentage(df)
```
