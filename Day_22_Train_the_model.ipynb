```python
df1 = df.copy()
```

```python
df1.isnull().sum()
```

```python
# fill null values with mean
numericColumns = ['Age','Years of Experience','Salary']

df1[numericColumns] = df1[numericColumns].fillna(df1[numericColumns].mean())
```

```python
df1.isnull().sum()
```

```python
# drop remaining null values - not important and less rows (2,3..)
df1 = df1.dropna()
df1.isnull().sum()
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()
```

```python
df1
```

```python
#Label Encoder
df1['Gender'] = le.fit_transform(df1['Gender'])
```

```python
df1['Education Level'] = le.fit_transform(df1['Education Level'])
```

```python
df1['Job Title'] = le.fit_transform(df1['Job Title'])
```

```python
df1
```

```python
# X, Y split,
x = df1.iloc[:,:-1]
```

```python
x
```

```python
y = df1.iloc[:,-1]
```

```python
y
```

```python
# train_test_split

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=17)
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
# Model Building

from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
# model evaluation
y_predict = lr.predict(x_test)
```

```python
error = y_predict - y_test
```

```python
error
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
mae = mean_absolute_error(y_test,y_predict)
```

```python
mse = mean_squared_error(y_test,y_predict)
```
