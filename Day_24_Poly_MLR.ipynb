```python

df = pd.DataFrame(californiaHousing.data, columns= californiaHousing.feature_names)
```

```python
df['Price'] = californiaHousing.target
```

```python
df = df.drop('y',axis=1)
```

```python
x = df[['MedInc','HouseAge','AveRooms']]
```

```python
x
```

```python
y = df[['Price']]
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=17)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_predict = lr.predict(x_test)
```

```python
y_predict
```

```python
error = y_test - y_predict
```

```python
error
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly = PolynomialFeatures(degree=2)
```

```python
x_poly = poly.fit_transform(x_train)
```

```python
x_poly
```

```python
x_train
```

```python
lr.fit(x_poly,y_train)
```

```python
result2 = poly.transform(x_test)
```

```python
y_predict2 = lr.predict(result2)
```

```python
y_predict2
```

```python
error2 = y_test - y_predict2
```
