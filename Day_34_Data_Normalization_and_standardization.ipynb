```python
df1 = df.copy()
```

```python
# drop last column
df1 = df1.iloc[:,:-1]
```

```python
# drop first column
df1 = df1.iloc[1:,:]
```

```python
df1.info()
```

```python
# remove unwanted columns
df1 = df1.drop(columns=['id'])
```

```python
df1['diagnosis'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df1['diagnosis'] = le.fit_transform(df1['diagnosis'])
```

```python
# y is first column, x is rest of the columns
y = df1.iloc[:,0]
x = df1.iloc[:,1:]
```

```python
x
```

```python
y
```

```python
from sklearn import preprocessing
norm_x = preprocessing.normalize(x)
```

```python
from sklearn.preprocessing import StandardScaler
std_datset = StandardScaler()
res = std_datset.fit_transform(x)
```

```python
norm_x
```

```python
res
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3, random_state=3)
```

```python
x_train_normal,x_test_normal,y_train_normal,y_test_normal = train_test_split(norm_x,y,test_size=0.3, random_state=3)
```

```python
x_train_scalar,x_test_scalar,y_train_scalar,y_test_scalar = train_test_split(res,y,test_size=0.3, random_state=3)
```

```python
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.datasets import make_classification
```

```python
model1 = SVC()
model2 = SVC()
model3 = SVC()
```

```python
model1.fit(x_train, y_train)
```

```python
model2.fit(x_train_normal, y_train_normal)
```

```python
model3.fit(x_train_scalar, y_train_scalar)
```

```python
y_predict = model1.predict(x_test)
```

```python
y_predict_normal = model2.predict(x_test_normal)
```

```python
y_predict_scalar = model3.predict(x_test_scalar)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
cm1 = confusion_matrix(y_test, y_predict)
```

```python
cm2 = confusion_matrix(y_test_normal, y_predict_normal)
```

```python
cm3 = confusion_matrix(y_test_scalar, y_predict_scalar)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns



# Define the labels for the confusion matrices
labels = ['M', 'B']

# Create a figure with three subplots
fig, axs = plt.subplots(1, 3, figsize=(18, 6))

# Plot the first confusion matrix
sns.heatmap(cm1, ax=axs[0], annot=True, cmap='Blues', fmt='d')
axs[0].set_title('1. Unprocessed Data')
axs[0].set_xlabel('Predicted Labels')
axs[0].set_ylabel('True Labels')
axs[0].set_xticks([0, 1])
axs[0].set_yticks([0, 1])
axs[0].set_xticklabels(labels)
axs[0].set_yticklabels(labels)

# Plot the second confusion matrix
sns.heatmap(cm2, ax=axs[1], annot=True, cmap='Blues', fmt='d')
axs[1].set_title('2. Normalization')
axs[1].set_xlabel('Predicted Labels')
axs[1].set_ylabel('True Labels')
axs[1].set_xticks([0, 1])
axs[1].set_yticks([0, 1])
axs[1].set_xticklabels(labels)
axs[1].set_yticklabels(labels)

# Plot the third confusion matrix
sns.heatmap(cm3, ax=axs[2], annot=True, cmap='Blues', fmt='d')
axs[2].set_title('3. Standardization')
axs[2].set_xlabel('Predicted Labels')
axs[2].set_ylabel('True Labels')
axs[2].set_xticks([0, 1])
axs[2].set_yticks([0, 1])
axs[2].set_xticklabels(labels)
axs[2].set_yticklabels(labels)

# Layout so plots do not overlap
fig.tight_layout()

plt.show()
```
