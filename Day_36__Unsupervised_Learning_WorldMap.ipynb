```python
X = df.drop('country', axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans

# Apply K-Means
kmeans = KMeans(n_clusters=3, random_state=42)  # Adjust n_clusters
df['Cluster'] = kmeans.fit_predict(X_scaled)
```

```python
import plotly.express as px
from scipy.cluster.hierarchy import linkage, fcluster
```

```python
# Perform hierarchical clustering
Z = linkage(X_scaled, method='ward')
df['Cluster'] = fcluster(Z, t=3, criterion='maxclust')  # Adjust 't' for cluster count

# Create the world map with Plotly
fig = px.choropleth(
    df,
    locations='country',            # Match countries in the dataset
    locationmode='country names',   # Match country names to map
    color='Cluster',                # Use clusters for color
    title='World Map of Clusters',
    color_continuous_scale='Viridis'
)

fig.update_layout(
    title='Hierarchical Clustering on World Map',
    title_x=0.5
)
fig.show()
```
