```python
df = df.drop('Id',axis=1)
```

```python
df = df.drop('Species', axis = 1)
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,0:4].values
```

```python
x
```

```python
from sklearn.cluster import KMeans
wcss = [] #within cluster sum of square
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', n_init=10, max_iter=300, random_state=0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1,11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('wcss')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
model2 = DBSCAN(eps=0.3, min_samples=10)
```

```python
model2.fit(x)
```

```python
label = model2.labels_
```

```python
label
```

```python
plt.scatter(x[:, 0], x[:, 1], c=label)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
model3 = AgglomerativeClustering(n_clusters=3)
```

```python
model3.fit(x)
```

```python
labels = model3.labels_
print(labels)
```

```python
# scatter plot

plt.scatter(x[:, 0], x[:, 1], c=labels)
```

```python
import scipy.cluster.hierarchy as sch
d_gram = sch.dendrogram(sch.linkage(x, method='ward'))
plt.title('Dendogram Method')
plt.xlabel('IRIS')
plt.ylabel('Euclidean Distances')
# plt.savefig('img_1.png', dpi=500)
plt.show()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score, davies_bouldin_score
```

```python
silhouette_score(x, labels)
```
